## Данные

Для анализа используется [IMDb Dataset](https://developer.imdb.com/non-commercial-datasets/), который состоит из нескольких файлов в формате TSV (сжатых в `.gz`). Каждый файл содержит информацию о фильмах, сериалах, эпизодах и участниках индустрии. Символ `\N` означает отсутствие или неизвестное значение.

**title.akas.tsv.gz** — альтернативные названия:

* `titleId` — уникальный идентификатор тайтла.
* `ordering` — порядковый номер записи для данного тайтла.
* `title` — локализованное название.
* `region` — регион показа.
* `language` — язык названия.
* `types` — тип альтернативного названия (например, dvd, festival, tv, video, original).
* `attributes` — дополнительные атрибуты.
* `isOriginalTitle` — индикатор оригинального названия (0 — нет, 1 — да).

**title.basics.tsv.gz** — базовая информация о тайтлах:

* `tconst` — уникальный идентификатор.
* `titleType` — тип (movie, short, tvSeries, tvEpisode и др.).
* `primaryTitle` — основное (наиболее распространённое) название.
* `originalTitle` — оригинальное название.
* `isAdult` — индикатор взрослого контента (0 — нет, 1 — да).
* `startYear` — год выхода (или начала показа сериала).
* `endYear` — год окончания сериала (для фильмов — `\N`).
* `runtimeMinutes` — продолжительность (в минутах).
* `genres` — до трёх жанров.

**title.crew.tsv.gz** — режиссёры и сценаристы:

* `tconst` — идентификатор тайтла.
* `directors` — список id режиссёров.
* `writers` — список id сценаристов.

**title.episode.tsv.gz** — информация об эпизодах сериалов:

* `tconst` — идентификатор эпизода.
* `parentTconst` — идентификатор сериала.
* `seasonNumber` — номер сезона.
* `episodeNumber` — номер эпизода.

**title.principals.tsv.gz** — ключевые участники проекта:

* `tconst` — идентификатор тайтла.
* `ordering` — порядковый номер записи для тайтла.
* `nconst` — идентификатор участника.
* `category` — категория (актёр, режиссёр и др.).
* `job` — профессия/роль в проекте (если указана).
* `characters` — персонажи (для актёров).

**title.ratings.tsv.gz** — рейтинги:

* `tconst` — идентификатор тайтла.
* `averageRating` — средний рейтинг (от 1 до 10).
* `numVotes` — количество голосов.

**name.basics.tsv.gz** — информация о людях:

* `nconst` — идентификатор участника.
* `primaryName` — основное имя (как в титрах).
* `birthYear` — год рождения.
* `deathYear` — год смерти (если применимо).
* `primaryProfession` — до трёх профессий (актёр, режиссёр и др.).
* `knownForTitles` — список тайтлов, по которым человек наиболее известен.

# 1. Подготовка данных

### Загрузка данных и знакомство с ними

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Загружаем все данные
opts = dict(sep='\t', na_values='\\N', compression='gzip', dtype=str)

basics = pd.read_csv('../data/title.basics.tsv.gz', **opts)
ratings = pd.read_csv('../data/title.ratings.tsv.gz', **opts)
crew = pd.read_csv('../data/title.crew.tsv.gz', **opts)
principals = pd.read_csv('../data/title.principals.tsv.gz', **opts)
names = pd.read_csv('../data/name.basics.tsv.gz', **opts)
akas = pd.read_csv('../data/title.akas.tsv.gz', **opts)
episode = pd.read_csv('../data/title.episode.tsv.gz', **opts)
```

```python
# Проверяем размеры загруженных таблиц
print("akas:", akas.shape)
print("basics:", basics.shape)
print("crew:", crew.shape)
print("episode:", episode.shape)
print("principals:", principals.shape)
print("ratings:", ratings.shape)
print("names:", names.shape)
```

**Вывод:** исходные данные представляют собой крупный и разнородный массив, требующий системной предобработки перед анализом.

---

### Приведение типов данных

Следующим шагом стало преобразование строковых значений в числовые там, где это необходимо:

* годы выпуска (`startYear`, `endYear`),
* длительность фильмов (`runtimeMinutes`),
* индикатор взрослого контента (`isAdult`),
* числовые показатели рейтингов и голосов (`averageRating`, `numVotes`),
* номера сезонов и эпизодов (`seasonNumber`, `episodeNumber`),
* годы рождения и смерти для персон (`birthYear`, `deathYear`).

Это позволило корректно использовать данные в математических операциях и статистическом анализе.

```python
basics["startYear"] = pd.to_numeric(basics["startYear"], errors="coerce").astype("Int64")
basics["endYear"] = pd.to_numeric(basics["endYear"], errors="coerce").astype("Int64")
basics["runtimeMinutes"] = pd.to_numeric(basics["runtimeMinutes"], errors="coerce").astype("Int64")
basics["isAdult"] = basics["isAdult"].astype("int8")
```

```python
ratings["averageRating"] = ratings["averageRating"].astype("float32")
ratings["numVotes"] = ratings["numVotes"].astype("int32")
```

```python
episode["seasonNumber"] = pd.to_numeric(episode["seasonNumber"], errors="coerce").astype("Int64")
episode["episodeNumber"] = pd.to_numeric(episode["episodeNumber"], errors="coerce").astype("Int64")
```

```python
names["birthYear"] = pd.to_numeric(names["birthYear"], errors="coerce").astype("Int64")
names["deathYear"] = pd.to_numeric(names["deathYear"], errors="coerce").astype("Int64")
```

**Вывод:** корректировка типов данных обеспечила основу для качественной аналитики, устранив проблему строковых значений в числовых полях.

---

### Проверка и удаление дубликатов

В каждой таблице были проверены дубликаты записей.

```python
print("Дубликаты в basics:", basics.duplicated().sum())
basics = basics.drop_duplicates()

print("Дубликаты в ratings:", ratings.duplicated().sum())
ratings = ratings.drop_duplicates()

print("Дубликаты в akas:", akas.duplicated().sum())
akas = akas.drop_duplicates()

print("Дубликаты в crew:", crew.duplicated().sum())
crew = crew.drop_duplicates()

print("Дубликаты в principals:", principals.duplicated().sum())
principals = principals.drop_duplicates()
```

```python
print("Будет удалено строк:", ratings[["averageRating", "numVotes"]].isna().any(axis=1).sum())
ratings.dropna(subset=["averageRating", "numVotes"], inplace=True)
```

**Вывод:** повторяющихся строк не было найдено, но для будущих обновлений датасета на всякий случай добавлено удаление дубликатов.

---

### Работа с пропущенными данными

* Пропущенные значения, отмеченные как `\N`, были преобразованы в `NaN` ещё на этапе загрузки данных.
* Для удобства произведена обработка колонок с множественными значениями. Значения для столбцов `genres` из таблицы **basics** и значения для столбцов `knownForTitles`, `primaryProfession` из таблицы **names** записанны строкой через запятую, преобразованы в список.

Это упростило работу с жанрами и подготовило данные для последующих визуализаций и статистических выборок.

```python
basics["genres"] = basics["genres"].apply(lambda x: x.split(",") if pd.notna(x) else [])
```

```python
names["knownForTitles"] = names["knownForTitles"].dropna().apply(lambda x: x.split(","))
names["primaryProfession"] = names["primaryProfession"].dropna().apply(lambda x: x.split(","))
```

**Вывод:** колонки, содержащие несколько значений в строке, преобразованы в список — это делает последующие объединения, фильтрации и подсчёты намного проще и надёжнее.

### Сохранение и анализ распределения типов тайтлов

```python
import os

# Создаём папку, если её нет
os.makedirs("data_processed", exist_ok=True)

# Сохраняем все обработанные таблицы
basics.to_parquet("data_processed/basics.parquet", index=False)
ratings.to_parquet("data_processed/ratings.parquet", index=False)
akas.to_parquet("data_processed/akas.parquet", index=False)
crew.to_parquet("data_processed/crew.parquet", index=False)
principals.to_parquet("data_processed/principals.parquet", index=False)
episode.to_parquet("data_processed/episode.parquet", index=False)
names.to_parquet("data_processed/names.parquet", index=False)
```

Этот формат оптимален для дальнейшей работы: он более эффективен по скорости чтения/записи и объёму памяти по сравнению с CSV. Таким образом, в любой момент можно вернуться к уже обработанным данным без повторного выполнения всех шагов предобработки.

### Фильтрация фильмов и сериалов

```python
movies = basics[basics["titleType"].isin(["movie", "tvMovie"])]
series = basics[basics["titleType"].isin(["tvSeries","tvMiniSeries"])]
```

Здесь из общей таблицы выделяются только объекты типа **фильмы** (`movie`, `tvMovie`) и **сериалы** (`tvSeries`, `tvMiniSeries`). Остальные типы (`short`, `videoGame`, `tvEpisode` и т. д.) для анализа отбрасываются, так как они не входят в основное исследование.


```python
movies = movies.merge(ratings, on="tconst", how="inner")
series = series.merge(ratings, on="tconst", how="inner")
print("Размер объединенной таблицы с фильмами:", movies.shape)
print("Размер объединенной таблицы с сериалами:", series.shape)
```

Фильмы и сериалы объединяются с таблицей `ratings` по ключу `tconst`. Используется `inner join`, чтобы оставить только те записи, у которых есть информация о рейтингах. После объединения выводятся размеры новых таблиц, чтобы оценить количество пригодных к анализу записей.

# 2. Тренды по жанрам

```python
movies_filtered = movies[movies["numVotes"] >= 1000]
print(movies_filtered.shape)
```

```python
series_filtered = series[series["numVotes"] >= 1000]
print(series_filtered.shape)
```

Мы оставили только те фильмы и сериалы, которые получили не менее 1000 голосов.
Это позволяет исключить «случайные» произведения с малым количеством оценок и сделать сравнение более надёжным.

```python
movies_exploded = movies_filtered.explode("genres").dropna(subset=["genres"])
```

```python
# Получаем список уникальных жанров
all_genres = movies_exploded["genres"].dropna().unique().tolist()

# Сортируем по алфавиту (удобнее для чтения)
all_genres = sorted(all_genres)

print("Количество жанров:", len(all_genres))
print(all_genres)
```

```python
# Считаем количество фильмов по жанрам
movies_exploded["genres"].value_counts()
```

```python
valid_genres = ["Drama", "Comedy", "Romance", "Action", "Crime", "Thriller", "Horror",
    "Documentary", "Adventure", "Mystery", "Family", "Fantasy", "Biography",
    "Sci-Fi", "History", "Music", "Animation", "War", "Musical", "Western", "Sport"]

movies_exploded = movies_exploded[movies_exploded["genres"].isin(valid_genres)]
```

```python
movies_exploded.to_parquet("data_processed/movies_exploded.parquet", index=False)
```

```python
import matplotlib.pyplot as plt

# Считаем количество фильмов по жанрам
genre_counts = movies_exploded["genres"].value_counts()

# Берем топ-10
top10_genres = genre_counts.head(10)

# Строим горизонтальный барчарт
plt.figure(figsize=(10, 6))
top10_genres.plot(kind="barh", color="skyblue")
plt.title("Топ-10 жанров по количеству фильмов (все годы)")
plt.xlabel("Количество фильмов")
plt.ylabel("Жанр")
plt.gca().invert_yaxis()  # чтобы самый популярный жанр был сверху
plt.show()
```

Лидируют Drama и Comedy — эти жанры формируют основной массив мировой киноиндустрии.
Драма — «универсальный жанр», который чаще всего используется как основа сюжетов.
Большая доля комедий отражает массовый запрос на лёгкий развлекательный контент.

```python
genre_counts = (
    movies_exploded
    .groupby(["startYear", "genres"])
    .size()
    .reset_index(name="count")
)
```

```python
genre_counts[genre_counts["startYear"] == 2024].sort_values("count", ascending=False)
```

```python
# Фильтруем только Drama
drama = movies_exploded[movies_exploded["genres"] == "Drama"]

# Считаем количество фильмов по годам
drama_by_year = drama.groupby("startYear").size()

# Строим график
plt.figure(figsize=(12, 6))
drama_by_year.plot(kind="line", color="darkred")
plt.title("Динамика количества фильмов жанра Drama по годам")
plt.xlabel("Год")
plt.ylabel("Количество фильмов")
plt.grid(True)
plt.show()
```

```python
# Фильтруем Drama
drama = movies_exploded[movies_exploded["genres"] == "Drama"]

# Фильтруем по годам 2000-2025
drama_recent = drama[(drama["startYear"] >= 2000) & (drama["startYear"] <= 2024)]

# Считаем количество фильмов по годам
drama_by_year = drama_recent.groupby("startYear").size()

# Строим график
plt.figure(figsize=(12, 6))
drama_by_year.plot(kind="line", color="darkred", marker="o")
plt.title("Динамика количества фильмов жанра Drama (2000–2025)")
plt.xlabel("Год")
plt.ylabel("Количество фильмов")
plt.xticks(range(2000, 2026, 2))  # метки каждые 2 года
plt.grid(True)
plt.show()
```

```python
# Фильтруем Horror
drama = movies_exploded[movies_exploded["genres"] == "Horror"]

# Фильтруем по годам 2000-2025
drama_recent = drama[(drama["startYear"] >= 2000) & (drama["startYear"] <= 2024)]

# Считаем количество фильмов по годам
drama_by_year = drama_recent.groupby("startYear").size()

# Строим график
plt.figure(figsize=(12, 6))
drama_by_year.plot(kind="line", color="darkred", marker="o")
plt.title("Динамика количества фильмов жанра Drama (2000–2025)")
plt.xlabel("Год")
plt.ylabel("Количество фильмов")
plt.xticks(range(2000, 2026, 2))  # метки каждые 2 года
plt.grid(True)
plt.show()
```

```python
# Средний рейтинг и медиана по жанрам
genre_stats = movies_exploded.groupby("genres")["averageRating"].agg(["mean", "median"]).sort_values("mean", ascending=False)
print(genre_stats)

# Boxplot — распределение рейтингов по жанрам
plt.figure(figsize=(12, 6))
sns.boxplot(x="averageRating", y="genres", data=movies_exploded, order=genre_stats.index)
plt.title("Распределение рейтингов фильмов по жанрам")
plt.xlabel("Средний рейтинг")
plt.ylabel("Жанр")
plt.show()
```

* Наибольшие средние рейтинги — у **Documentary**, **Biography** и **History**.
* Массовые жанры (Comedy, Horror, Action) в среднем оцениваются ниже.

* Можно предположить, что зрители склонны выставлять умеренные оценки, а экстремальные значения чаще связаны либо с узкой фанатской аудиторией, либо с заведомо провальными проектами.

### Отличия между фильмами и сериалами.

```python
# --- Статистика для фильмов ---
stats_movies = {
    "Среднее": movies_filtered["averageRating"].mean(),
    "Медиана": movies_filtered["averageRating"].median(),
    "Мода": movies_filtered["averageRating"].mode().iloc[0],
    "Стандартное отклонение": movies_filtered["averageRating"].std()
}

# --- Статистика для сериалов ---
stats_series = {
    "Среднее": series_filtered["averageRating"].mean(),
    "Медиана": series_filtered["averageRating"].median(),
    "Мода": series_filtered["averageRating"].mode().iloc[0],
    "Стандартное отклонение": series_filtered["averageRating"].std()
}

print("Фильмы:\n", stats_movies)
print("\nСериалы:\n", stats_series)
```

Сравнение по основным метрикам:
- Средний рейтинг фильмов ≈ 6.23
- Средний рейтинг сериалов ≈ 7.2  

```python
plt.figure(figsize=(8,6))
plt.boxplot(
    [movies_filtered["averageRating"], series_filtered["averageRating"]],
    labels=["Фильмы", "Сериалы"],
    patch_artist=True
)
plt.ylabel("Средний рейтинг")
plt.title("Сравнение распределения рейтингов")
plt.show()
```

Boxplot показывает, что распределения в целом схожи, но сериалы чаще имеют более высокие медианные значения и больший разброс в верхней части шкалы.

### Проверка статистической значимости: t-тест, тест Манна-Уитни

```python
# --- Статистические тесты ---
t_stat, t_p = stats.ttest_ind(
    movies_filtered["averageRating"], 
    series_filtered["averageRating"], 
    equal_var=False
)

u_stat, u_p = stats.mannwhitneyu(
    movies_filtered["averageRating"], 
    series_filtered["averageRating"], 
    alternative="two-sided"
)

print("\nT-тест: статистика =", t_stat, ", p-value =", t_p)
print("Манна–Уитни: статистика =", u_stat, ", p-value =", u_p)
```

Проведённые тесты (t-test и Манна–Уитни) показали, что различия между средними рейтингами фильмов и сериалов статистически значимы (p < 0.05).  
Это означает, что более высокие рейтинги сериалов — не случайность, а устойчивая тенденция.

```python
avg_by_genre = exploded_focus.groupby(["startYear", "genres"])["runtimeMinutes"].mean().reset_index()
```

```python
plt.figure(figsize=(14,7))

for genre in focus_genres:
    subset = avg_by_genre[avg_by_genre["genres"] == genre].copy()
    if not subset.empty:
        subset = subset.sort_values("startYear")
        subset["smooth"] = subset["runtimeMinutes"].rolling(window=5, center=True).mean()
        plt.plot(subset["startYear"], subset["smooth"], label=genre, linewidth=2)

plt.title("Средняя длительность фильмов по жанрам (сглаживание 5 лет)")
plt.xlabel("Год")
plt.ylabel("Средняя длительность (минуты)")
plt.legend(title="Жанры", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Сравнение пяти жанров показало:

* Drama — как правило, самые длинные фильмы (средняя длительность ~110 мин).
* Comedy — самые короткие из популярных жанров (~95 мин).
* Action — стабильно держатся в районе 105–110 мин, с ростом в последние десятилетия.
* Horror — почти всегда короче среднего (~90–95 мин).
* Animation — в целом компактные (80–90 мин), без заметной тенденции к увеличению.

На графиках видно, что различия между жанрами устойчивы во времени: комедии и анимация остаются короткими, драмы и боевики длиннее.
